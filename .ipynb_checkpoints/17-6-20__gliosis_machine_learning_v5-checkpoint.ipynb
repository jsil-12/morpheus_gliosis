# Mouse Class

```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from mpl_toolkits.mplot3d import Axes3D
from collections import OrderedDict
#import hdbscan
from sklearn.preprocessing import StandardScaler
import os 
import copy


## t-SNE
#boxsize = 150
class mouse:
    def __init__(self,ID,d,cond,sID,file,mBox=100):
        self.day = d
        self.ID = ID
        self.sections={}
        self.add_section(sID,cond,file,box=mBox)
        self.summary_data = None
    
    def add_section(self,sID,cond,file,box=100):
        if sID in self.sections:
            self.sections[sID].add_section(cond, pd.read_csv(file, delimiter="\t"))
        else:
            self.sections[sID] = section(sID, cond, pd.read_csv(file, delimiter="\t"),sBox=box)
            
    def append_section(self,sID,cond,file,box=100):
        self.sections[sID].data[cond] = pd.concat([self.sections[sID].data[cond],pd.read_csv(file, delimiter="\t")],ignore_index=True)
        
    def get_sections(self):
        return self.sections
        
    def update_section(self,sec):
        sID = sec.sID
        self.sections[sID] = sec

class section:
    def __init__(self,sID,cond,pd,sBox=125):
        self.boxsize = sBox
        self.sID = sID
        self.data = {}
        self.add_section(cond,pd)
        self.scale = self.set_scale()
        
    #adds a section, when both FUS and nonFUS sides are added. 
    #the section is thresholded. Threshold = noFUS mean + 2SD.
    #the thresholded section is then clustered
    def add_section(self, cond, pd):
        self.data[cond] = pd
        
    def set_scale(self):
        IDs = list(self.data.keys())
        data = self.data[IDs[0]]
        scale = float((self.boxsize) / data["Width"].iloc[5]) #note: .iloc[5] is always a box
        return scale

def add_mice(d,b=125,segment=False):
    mice = {}
    files = os.listdir(d)
    #print(files)
    for f in files:
        print(f)
        ms, day, cond, cell, sect, start = f.split("_")
        if sect.endswith(".tif.txt"):
            sect = sect[:-8]
        if cell in mice:
            if day in mice[cell]:
                if ms in mice[cell][day]:
                    #addSection(self,sID,cond,file)
                    if sect in mice[cell][day][ms].sections:
                        if cond in mice[cell][day][ms].sections[sect].data:
                            mice[cell][day][ms].append_section(sect,cond,d+f,box=b)
                        else:
                            mice[cell][day][ms].add_section(sect,cond,d+f,box=b)
                    else:
                        mice[cell][day][ms].add_section(sect,cond,d+f,box=b)
                else:
                    mice[cell][day][ms] = mouse(ms,day,cond,sect,d+f,mBox=b)
            else:
                mice[cell][day] = {}
                mice[cell][day][ms] = mouse(ms,day,cond,sect,d+f,mBox=b)
        else:
            mice[cell]={}
            mice[cell][day]={}
            mice[cell][day][ms]=mouse(ms,day,cond,sect,d+f,mBox=b)
    return mice

def get_mouse_sections(mice,cell,day,ID):
    return mice[cell][day][ID].sections

def get_ms_sections(ms):
    return ms.sections

def iter_mice(mice,cell):
    for day in mice[cell]:
            for ID in mice[cell][day]:
                yield mice[cell][day][ID]

def iter_sections(ms,cond="All"):
    if cond == "All":
        yield ms.sections
    else:
        yield ms.sections[cond]
        

           
```

# _PCA_ _Novel_Discovery_ _DBSCAN_

```python
from sklearn.decomposition import PCA
from sklearn import svm
from sklearn.cluster import DBSCAN
import math
import collections
import bisect

#assigns the components to a dataframe
def assign_components(X,Xt):
    X=X.copy()
    for i in range(0,len(Xt.T)):
        col="comp"+str(i+1)
        X[col]=Xt[:,i]
    return X

#input a section, outputs PCA transformed nf and f data in the section
def PCA_section(sec,nComps=3,features=["Mean","Area","Circ.","IntDen","%Area","Median","Skew","Solidity","StdDev"],control="noFUS",treatment="FUS"):
    X=copy.deepcopy(sec)
    nf=X.data[control]
    f=X.data[treatment]
    pca=PCA(n_components=nComps).fit(nf[features])
    nf_t = pca.transform(nf[features])
    f_t = pca.transform(f[features])
    X.data[control] = assign_components(nf,nf_t)
    X.data[treatment] = assign_components(f,f_t)
    return X

def novelty_detection(sec,_nu=0.5,_gamma='auto',_kernel='rbf',features=["Mean","Area","Circ.","IntDen","%Area","Median","Skew","Solidity","StdDev"],train_self=False,report=False,control="noFUS",treatment="FUS"):
    reports={}
    sec=copy.deepcopy(sec)
    print(_nu)
    if train_self:
        estimator = svm.OneClassSVM(nu=_nu,gamma=_gamma,kernel=_kernel)
        for cond in sec.data:
            x=sec.data[cond]
            estimator.fit(x[features])
            test = estimator.predict(x[features])
            x["self_outlier"] = test
            if report:
                reports[cond]=report_percent_outliers(X,cond,outlier="self_outlier")
            sec.data[cond] = x
    else:
        estimator = svm.OneClassSVM(nu=_nu,gamma=_gamma,kernel=_kernel)
        X_train=sec.data[control]
        X_test=sec.data[treatment]
        estimator.fit(X_train[features])
        y_pred_train = estimator.predict(X_train[features])
        y_pred_test = estimator.predict(X_test[features])
        X_train["outlier"]=y_pred_train
        X_test["outlier"]=y_pred_test
        if report:
            reports[control]=[report_percent_outliers(X_train,control)]
            reports[treatment]=[report_percent_outliers(X_test,treatment)]
        sec.data[control] = X_train
        sec.data[treatment] = X_test
    if report:
        return sec,reports
    else:
        return sec

def report_percent_outliers(X,cond,outlier="outlier"):
    perc=len(X.loc[X[outlier]==-1,X.columns.values])/len(X)
    print("condition: %s, percent outliers %.3f" %(cond,perc))
    return perc

def trim_section(sec,features=["Mean","Area","Circ."],greater=[True,True,True],nf_threshold=[0,0,0],f_threshold=[0,0,0],outlier_only=True,control="noFUS",treatment="FUS"):
    X = copy.deepcopy(sec)
    f=X.data[treatment]
    nf=X.data[control]
    if outlier_only:
        f = f.loc[f["outlier"]==-1,f.columns.values]
        nf = nf.loc[nf["outlier"]==-1,nf.columns.values]
    for i in range(0,len(features)):
        if greater[i] == True:
            print(features)
            f = f.loc[f[features[i]] >= f_threshold[i], f.columns.values]
            nf = nf.loc[nf[features[i]] >= nf_threshold[i], nf.columns.values]
        else:
            f = f.loc[f[features[i]]  <= f_threshold[i], f.columns.values]
            nf = nf.loc[nf[features[i]]  <= nf_threshold[i], nf.columns.values]
    X.data[control]=nf
    X.data[treatment]=f
    return X

def threshold_section(sec,measure="Mean",control="noFUS",treatment="FUS"):
    X = copy.deepcopy(sec)
    nf = X.data[control]
    f = X.data[treatment]
    nf = nf.loc[nf[measure]>2,nf.columns.values]
    f = f.loc[f[measure]>2,f.columns.values]
    X.data[control] = nf
    X.data[treatment] = f
    return X

def density_cluster(sec,_eps=None,_min=10,components=["BX","BY"],cl_name="cluster",outliers=None,control="noFUS",treatment="FUS"):
    boxsize = sec.boxsize
    if _eps is None:
        print("accessed")
        _eps = math.sqrt((2.0*boxsize**2)+1)/1.5
    print(_eps)
    X=copy.deepcopy(sec)
    nf = X.data[control]
    f = X.data[treatment]
    if (outliers==None) == False:
        #nf_rest = nf.loc[nf[outliers]==1,nf.columns.values]
        #f_rest = f.loc[f[outliers]==1,f.columns.values]
        #nf_rest[cl_name] = [-1]*len(nf_rest)  #if not an outlier, then it is -1 cluster
        #f_rest[cl_name] = [-1]*len(f_rest) #if not an outlier, then it is -1 cluster
        
        nf = nf.loc[nf[outliers]==-1,nf.columns.values]
        f = f.loc[f[outliers]==-1,f.columns.values]
    clusterer=DBSCAN(min_samples=_min,eps=_eps)
    
    # cluster noFUS + FUS
    if len(nf) > 0:
        nf[cl_name]=clusterer.fit_predict(nf[components])
    else:
        nf = X.data[control]
        nf[cl_name] = [-1]*len(X.data[control])
    if len(f) > 0:
        f[cl_name]=clusterer.fit_predict(f[components])
    else:
        f = X.data[treatment]
        f[cl_name] = [-1]*len(X.data[treatment])
        
    #restore removed boxes
    #if (outliers==None) == False:
        #nf = pd.concat([nf, nf_rest])
        #f = pd.concat([f,f_rest])
    
    X.data[control] = nf
    X.data[treatment] = f
    return X


    
def get_clustered(sec):
    sec = copy.deepcopy(sec)
    for c in sec.data:
        data = sec.data[c]
        sec.data[c] = data.loc[data["cluster"]>-1,data.columns.values]
    return sec

#returns a section with focal clusters removed from the proximal clusters
def subtract_section(prox_sec,foc_sec,clusters=True,write=False,mID="JS1",day="1D",cell="microglia",path="C:\\Users\\joey\\Documents\\LabWork\\Staining\\immunofluorescense\\Gfap_Iba1_Ki67\\GFAP_Iba1_ki67_batch_output\\input_cluster_files",control="noFUS",treatment="FUS"):
    subtracted = copy.deepcopy(prox_sec)
    for cond in [treatment,control]:
        proximal = prox_sec.data[cond]
        focal = foc_sec.data[cond]
        if clusters:
            proximal = proximal.loc[proximal["cluster"]>-1,proximal.columns.values]
            focal = focal.loc[focal["focal_cluster"]>-1,focal.columns.values]
            to_remove=[]
                        
            #box_length=prox_sec.boxsize/prox_sec.scale
            #remove_box = proximal.loc[]
            #df.drop(df.index[[1,3]])
            #subtracted.data[cond] = subtracted.data[cond].drop(subtracted.data[cond].index[to_remove])
                
            
            
            
            for box in focal[["BX","BY"]].values:
                box_length = prox_sec.boxsize/prox_sec.scale
                #R = cl_DF.loc[(cl_DF["lowX"]<=x["X"])&(cl_DF["highX"]>x["X"])&(cl_DF["lowY"]<=x["Y"])&(cl_DF["highY"]>x["Y"]),cl_DF.columns.values]
                
                #remove inner boxes + outer boxes
                box_to_remove = proximal.loc[(proximal["BX"] >= box[0]-(box_length/2)) & (proximal["BX"] < box[0]+box_length) & (proximal["BY"] >= box[1]-(box_length/2)) & (proximal["BY"] < box[1]+box_length), proximal.columns.values]
                #to_remove.append(box_to_remove)
                subtracted.data[cond] = subtracted.data[cond].loc[~(subtracted.data[cond].index.isin(box_to_remove.index)), subtracted.data[cond].columns.values]
            
            
            
            '''
            box_length = prox_sec.boxsize/prox_sec.scale
            #print(proximal.index)
            for i in proximal.index:
                #print(i)
                box=proximal[["BX","BY"]].ix[i]
                #print(box)
                for fbox in focal[["BX","BY"]].values:
                    if ( (box[0] <= fbox[0]) and ( (box[0]+ box_length) > fbox[0]) and  (box[1] <= fbox[1]) and ( (box[1]+ box_length) > fbox[1]) ):
                        to_remove.append(i)
            
            if to_remove:
                subtracted.data[cond] = subtracted.data[cond].loc[~(subtracted.data[cond].index.isin(to_remove)),subtracted.data[cond].columns.values]
            '''
            
        #print("proximal: %s, focal: %s, to subtract: %s" %(len(proximal),len(focal),len(np.unique(to_remove))))
        print("proximal: %s, focal: %s, subtracted: %s" %(len(proximal),len(focal),len(subtracted.data[cond].loc[subtracted.data[cond]["cluster"]>-1,subtracted.data[cond].columns.values])))
    if write:
        write_clusters(subtracted,mID,day,cell,path=path,clust_type="subtracted",clust_name="cluster")
    return subtracted
    

def combine_section(prox_sec,foc_sec,clusters=True,write=False,mID="JS1",day="1D",cell="microglia",path="C:\\Users\\joey\\Documents\\LabWork\\Staining\\immunofluorescense\\Gfap_Iba1_Ki67\\GFAP_Iba1_ki67_batch_output\\input_cluster_files",control="noFUS",treatment="FUS"):
    combined = copy.deepcopy(prox_sec)
    for cond in [treatment,control]:
        proximal = prox_sec.data[cond]
        focal = foc_sec.data[cond]
        if clusters:
            proximal = proximal.loc[proximal["cluster"]>-1,["BX","BY","cluster"]]
            focal = focal.loc[focal["focal_cluster"]>-1,["BX","BY","focal_cluster"]]
            
            focal = focal.rename(columns={'focal_cluster':'cluster'})
        print(focal)
        print(proximal)
        comb = pd.concat([proximal,focal])
        comb = comb.reset_index().drop_duplicates(subset='index', keep='last').set_index('index')
        #combined.data[cond] = proximal.append(focal)
        #combined.data[cond] = combined.data[cond].reset_index().drop_duplicates(subset='index', keep='last').set_index('index')
        combined.data[cond] = comb
        #print(subtracted.data[cond])
        print("proximal: %s, focal: %s, combined: %s" %(len(proximal),len(focal),len(combined.data[cond])))
    if write:
        write_clusters(combined,mID,day,cell,path=path,clust_type="combined",clust_name="cluster")
    return combined

#converts a section FUS and noFUS clusters and writes them as a string to a file
def write_clusters(sec,mID,day,cell,path="C:\\Users\\joey\\Documents\\LabWork\\Staining\\immunofluorescense\\Gfap_Iba1_Ki67\\GFAP_Iba1_ki67_batch_output\\input_cluster_files\\",clust_type="proximal",clust_name="cluster"):
    for cond in sec.data:
        filename=mID+"_"+day+"_"+cond+"_"+cell+"_"+sec.sID+".txt"
        toWrite=""
        X = sec.data[cond]
        Xt = X.loc[X[clust_name]>-1,["BX","BY"]]
        Xt["BX"] = Xt["BX"]*sec.scale
        Xt["BY"] = Xt["BY"]*sec.scale
        print(sec.scale)
        for i in range(0,len(Xt)):
            toWrite+=str(Xt.iloc[i]["BX"])+"-"
            toWrite+=str(Xt.iloc[i]["BY"])+"-"
            toWrite+=str(sec.boxsize)+"="
        toWrite=toWrite[:-1] #get rid of last =
        with open(path+"\\"+cell+"\\"+clust_type+"\\"+filename,'w') as w:
            w.write(toWrite)

print("done")
```

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scipy
from scipy.stats import ks_2samp

def get_focal_threshold(X,measure="Mean",plot=False,cluster_only=False,derivative=False,fit_curve=True,scan=False,max_curve=False):
    Xt = copy.copy(X)
    if cluster_only:
        print("only clusters!")
        Xt = Xt.loc[Xt["cluster"]>-1, Xt.columns.values]
        print("cluster length %s" %(len(Xt)))
        print(len(X))
    #S = f[measure].sort()
    Xt = Xt.sort_values(by=measure,ascending=True)
    if len(Xt) > 0:
        if derivative:
            slopes = get_derivative(Xt[measure].values)
            slope_deriv = get_derivative(slopes)
            Ys = slope_deriv
            Xs = list(range(0,len(Ys)))
            y_min = min(list(Ys))
            X_min = list(Ys).index(y_min)
            Xcut = np.array(Ys[X_min:])
            X_minT = X_min
            Xzero = np.argmax(Xcut >= 0)+X_min
            
            Xthresh = Xzero*4
            thresh = Xt[measure].iloc[Xthresh]
            Yv = Xt[measure].values
            Xv = range(0,len(Yv))
            print("threshold value: %s" %(thresh))
            if plot:
                plt.plot(Xs,Ys)
                plt.axvline(x=Xzero)
                plt.show()

                plt.plot(Xv,Yv)
                plt.axvline(x=Xthresh)
                plt.show()
            return thresh
        elif fit_curve:
            fit_curves(Xt[measure].values)
        else:            
            thresh = find_optimal_elbow(Xt[measure],converge=False,scan=scan,num_iters=3,penalty=0.025,plot=plot)
            print("#########################")
            print("elbow point: (%s)" %(thresh))
            print("#########################")
            if plot:
                Ys = Xt[measure].values
                Xs = range(1,len(Ys)+1)
                plt.plot(Xs,Ys)
                if not (thresh == 99):
                    plt.axhline(y=thresh)
                plt.show()
            return thresh
    else:
        return max(X[measure]) #if no focal threshold -- due to no clusters, return max threshold value to ensure no subsequent clustering

#input a list
def find_elbow(curve):
    nPoints = len(curve)
    allCoord = np.vstack((range(nPoints), curve)).T
    np.array([range(nPoints), curve])
    firstPoint = allCoord[0]
    lineVec = allCoord[-1] - allCoord[0]
    lineVecNorm = lineVec / np.sqrt(np.sum(lineVec**2))   #b^
    vecFromFirst = allCoord - firstPoint
    scalarProduct = np.sum(vecFromFirst * np.matlib.repmat(lineVecNorm, nPoints, 1), axis=1)
    
    vecFromFirstParallel = np.outer(scalarProduct, lineVecNorm)
    distFromFirstParallel = np.sqrt(np.sum(vecFromFirstParallel ** 2, axis=1))
    
    vecToLine = vecFromFirst - vecFromFirstParallel
    distToLine = np.sqrt(np.sum(vecToLine ** 2, axis=1))
    idxOfBestPoint = np.argmax(distToLine)
    angle = np.arctan(distToLine[idxOfBestPoint]/distFromFirstParallel[idxOfBestPoint])
    print("elbow point: %s, parallel dist: %s, perpendicular dist: %s, angle: %s" %(idxOfBestPoint,distFromFirstParallel[idxOfBestPoint],distToLine[idxOfBestPoint],angle))
    return [idxOfBestPoint,curve[idxOfBestPoint],distToLine[idxOfBestPoint]]

def find_optimal_elbow(curve,_min=False,converge=False,num_iters=1,penalty=0.05,scan=False,plot=False,max_curve=False):
    if converge:
        for i in range(0,num_iters):
            N = int(len(curve)*penalty*i)
            fo = iter_elbow(curve[N:],num_iters=1)
            reverse_curve = curve.sort_values(ascending=False)
            rev = iter_elbow(reverse_curve[N:],num_iters=1)
            if (fo[1]/rev[1] > 0.95) & (rev[1]/fo[1] > 0.95):
                P = (fo[0]+int(len(curve)*penalty*i),fo[1])
                return P[1]
            else:
                return (len(curve),max(curve))[1]
    elif scan:
        T = scan_elbow(curve,penalty=penalty,plot=plot,max_curve=max_curve)
        if T == -1:
            T = 99
        return T
    else:
        print("iter elbow normal")
        return iter_elbow(curve,penalty=penalty,num_iters=num_iters)[1]

def iter_elbow(curve,num_iters=3,penalty=0.05):
    if num_iters > 1:
        result=(len(curve),max(curve))
        print("itering for %s rounds" %(num_iters))
        for i in range(0,num_iters):
            N = int(len(curve)*penalty*i)
            N_1 = int(len(curve)*penalty*(i+1))
            R = find_elbow(curve.iloc[N:].values)
            R_1 = find_elbow(curve.iloc[N_1:].values)
            print(R_1[1]/R[1])
            print((R[1]/R_1[1]))
            if ( ((R_1[1]/R[1]) > 0.95) & ((R[1]/R_1[1]) > 0.95) ):
                print("Acessed!")
                R = (R[0] + int(len(curve)*penalty*i),R[1])
                result = R
        return result
    else:
        print("no curve iteration")
        print(find_elbow(curve))
        return find_elbow(curve)
    
def scan_elbow(curve,penalty=0.05,plot=False,min_curve_distance=0.5,max_curve=False):
    sub = int(penalty * len(curve))
    if sub == 0:
        sub = 1
    nIters = int(len(curve)/sub)
    elbows = []
    curve_dist = []
    for i in range(0,nIters):
        start = sub*i
        elbow = find_elbow(curve.iloc[start:].values)
        elbows.append(elbow[1])
        curve_dist.append(elbow[2])
    if plot:
        plt.plot(range(0,nIters),elbows,'b')
        print("-----------")
        print(curve_dist)
        print("-----------")
        
        plt.plot(range(0,nIters),curve_dist,'r')
        plt.show()
    thresh = scipy.stats.mode(elbows).mode[0]
    threshindex = np.where(elbows==thresh)[0][0] # get first index where mode elbow point i.e. threshold value is reached.
    print(threshindex)
    if(curve_dist[threshindex]<min_curve_distance): # assess curvature at this point
        thresh = -1
    return thresh

def get_derivative(curve):
    slope = []
    step=2
    for i in range(0,len(curve)-1,step):
        slope.append((curve[i+1]-curve[i])/5)
    return slope

#input a mouse, returns clustered datapoints from all sections of a condition.
def merge_outliers(ms, cond="FUS",ignore_index=True):
    toMerge=[]
    for s in ms.sections:
        sec = ms.sections[s].data[cond]
        toMerge.append(sec.loc[sec["outlier"]==-1,sec.columns.values])
    merged = pd.concat(toMerge,ignore_index=ignore_index)
    return merged

def merge_outliers_for_day(mice,day="1D",cond="FUS",cell="microglia",ignore_index=True):
    toMerge=[]
    for ms in mice[cell][day]:
        toMerge.append(merge_clusters(mice[cell][day][ms]))
    merged = pd.concat(toMerge,ignore_index=ignore_index)
    return merged

def find_focal_threshold(mice,day="1D",cell="microglia",measure="Median",plots=False,cluster_only=True):
    if day == "All":
        threshold = {}
        for d in ["1D","4D","7D","10D"]:
            outliers = merge_outliers_for_day(mice,day=d,cell=cell,cond="FUS")
            threshold[d]=get_focal_threshold(outliers,measure=measure,plot=plots,cluster_only=cluster_only)
        return threshold
    else:
        outliers = merge_outliers_for_day(mice,day=day,cell=cell,cond="FUS")
        return get_focal_threshold(outliers,measure=measure,plot=plots,cluster_only=cluster_only)
    
def find_focal_threshold_per_mouse(mice,day="1D",cell="microglia",cond="FUS",measure="Mean",cluster_only=True,plots=False):
    threshold={}
    for ms in mice[cell][day]:
        merged = merge_outliers(mice[cell][day][ms],cond=cond)
        threshold[ms] = get_focal_threshold(merged,measure=measure,plot=plots,cluster_only=cluster_only)
    return threshold
    
def line(x, a, b):
    return a * x + b

def exp_curve(x,a,b,c):
    #print("compontents: a:%s, b:%s, c:%s, x:%s" %(a,b,c,x))
    #return a*(b**x)
    #a*np.exp(-b*x)+c
    #return a*np.exp(b*x)+c
    return a*(b**x) + c

def simple_exp_curve(x,a,b,c):
    #return a*np.exp(b*x)
    return a*(b**x)+c

def second_order_poly(x,a,b,c):
    return a*x**2 + b*x + c

def third_order_poly(x,a,b,c,d):
    return a*x**3+b*x**2+c*x+d

def fourth_order_poly(x,a,b,c,d,e):
    return a*x**4+b*x**3+c*x**2+d*x+e
    
def fit_curves(curve):
    x = np.linspace(0., len(curve), len(curve))
    y = copy.deepcopy(curve)
    #print(y)
    #print(y.loc[y=="Nan",y.columns.values])
    
    popt, pcov = curve_fit(line, x, y)
    popt_s, pcov_s = curve_fit(simple_exp_curve,x,y)
    #popt_exp, pcov_exp = curve_fit(exp_curve,x,y)
    #popt_2poly, pcov_2poly = curve_fit(second_order_poly,x,y)
    
    #print("###")
    #print(line(x,*popt))
    #popt_poly, pcov_poly = curve_fit(third_order_poly,x,y)
    #popt_exp, pcov_exp = curve_fit(simple_exp_curve,x,y)
    
    '''
    popt_est, pcov_est = curve_fit(simp_exp_func,x,y)
    init_params = np.append(popt_est,0)
    popt_exp, pcov_exp = curve_fit(exp_func, x, y)#0.43637567  0.02932099  #,p0=init_params
    '''
    #print("###")
    #print(exp_func(x,*popt_exp))
    #print("###")
    #print(popt_exp)
    #xfine_exp = np.linspace(0., 100., 100)
    
    plt.plot(x, simple_exp_curve(x, *popt_s), 'r')
    #plt.plot(x,second_order_poly(x,*popt_2poly),'r')
    #plt.plot(x,third_order_poly(x,*popt_poly),'y')
    plt.plot(x, line(x, *popt), 'g')
    plt.plot(x, y, '.')
    plt.show()

    print(ks_2samp(y, line(x,*popt)))
    print(ks_2samp(y, simple_exp_curve(x,*popt_s)))
    
    print("linear covariate: %s, exponent covariate: %s" %(pcov,pcov_s))
    #print(ks_2samp(y, second_order_poly(x,*popt_2poly)))
    #print(ks_2samp(y, third_order_poly(x,*popt_poly)))

    

#focal_thresholds = find_focal_threshold(proximal,plots=True,day="All",measure="Mean")
#print(focal_thresholds)
    
```

```python
import os
import pandas as pd

#read an XML and return a dictionary of markers : [coordinates]
def parse_xml(file,scale=1.5):
    data = {}
    with open(file,'r') as R:
        lines = R.readlines(-1)
        lnum=0
        while lnum < len(lines):
            line  = lines[lnum].strip()
            if line == "<Marker_Type>":
                marker = lines[lnum+1].strip().strip("</Type>")
                data[marker] = []
                while not (line == "</Marker_Type>"):
                    if line == "<Marker>":
                        x=lines[lnum+1].strip().strip("</MarkerX>")
                        y=lines[lnum+2].strip().strip("</MarkerY>")
                        data[marker].append([float(x)/scale,float(y)/scale])
                    lnum+=1
                    line  = lines[lnum].strip()
            lnum+=1
    return data

def xml_to_df(file,scale=1.5,targets=["1","2"]):
    toDF = {}
    data = parse_xml(file,scale=scale)
    #print(data[targets[1]])
    #print(list(zip(*data[targets[1]])))
    for target in targets:
        if len(data[target]) > 0:
            df = pd.DataFrame(list(zip(*data[target]))).T
            #print(df)
            df.columns = ["X","Y"]
            #print(df)
            toDF[target] = df        
            #df = pd.DataFrame(data[target],columns=["X","Y"])
            #toDF[target] = df
    return toDF

def check_all_cluster_files(cellpath="G:\\lab_files\\cell_counts\\best\\XMLs\\", clusterpath="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\input_cluster_files\\novel_outlier\\microglia\\combined\\",cell="microglia",target="1",scale=1.5):
    cellfiles = os.listdir(cellpath)
    counts = {"noFUS":{},
             "FUS":{}}
    for file in cellfiles:
        #print(file)
        header,ID,day,cond,sec = file.split("_")
        sec = sec[:-4] #strip .txt
        clfile = ID+"_"+day+"_"+cond+"_"+cell+"_"+sec+".txt"
        clusters = open_cluster_file(clusterpath+clfile)
        #print(clusters.sort_values("BX"))
        cells = xml_to_df(cellpath+file,scale=scale)
        incl, outcl = check_in_cluster(cells,clusters,target=target)
        print("file: %s, incluster: %s, outcluster:%s" %(file,incl,outcl)) 
        if day in counts[cond]:
            if ID in counts[cond][day]:
                counts[cond][day][ID]["in_cluster"] += incl
                counts[cond][day][ID]["out_cluster"] += outcl
            else:
                counts[cond][day][ID] = {}
                counts[cond][day][ID]["in_cluster"] = incl
                counts[cond][day][ID]["out_cluster"] = outcl
        else:
            counts[cond][day] = {}
            counts[cond][day][ID] = {}
            counts[cond][day][ID]["in_cluster"] = incl
            counts[cond][day][ID]["out_cluster"] = outcl
    return counts
        
def open_cluster_file(file):
    with open(file) as f:
        X = []
        Y = []
        for line in f:
            boxes = line.split("=")
            for box in boxes:
                x,y,size = box.split("-")
                X.append(x)
                Y.append(y)
        df = pd.DataFrame(list(zip(X,Y)),columns=["BX","BY"])
        df = df.apply(pd.to_numeric)
        #print(df)
        return df
            
def check_in_cluster(cells,clusters,target,blength = 150):
    incl = 0
    outcl = 0
    data = clusters
    #print(cells)
    if target in cells:
        prolif = cells[target][["X","Y"]]
        if len(prolif) > 0:
            #if len(data) > 0:
            for index, row in prolif.iterrows():
                c = row
                #print(c["X"])
                #print(c["Y"])
                indexes = data.loc[((data["BX"]<c["X"])
                                   & ((data["BX"] + blength) >= c["X"])
                                   & (data["BY"]<c["Y"])
                                   & ((data["BY"] + blength) >= c["Y"])
                                   )].index
                if len(indexes) > 0:
                    #print("IN!!!")
                    #2print("INDEX")
                    incl += 1
                else:
                    outcl += 1
    return incl, outcl


check_all_cluster_files(scale=1.0,target="1",cell="microglia")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab

#read an XML and return a dictionary of markers : [coordinates]
def parse_xml(file,scale=1.5):
    data = {}
    with open(file,'r') as R:
        lines = R.readlines(-1)
        lnum=0
        while lnum < len(lines):
            line  = lines[lnum].strip()
            if line == "<Marker_Type>":
                marker = lines[lnum+1].strip().strip("</Type>")
                data[marker] = []
                while not (line == "</Marker_Type>"):
                    if line == "<Marker>":
                        x=lines[lnum+1].strip().strip("</MarkerX>")
                        y=lines[lnum+2].strip().strip("</MarkerY>")
                        data[marker].append([float(x)/scale,float(y)/scale])
                    lnum+=1
                    line  = lines[lnum].strip()
            lnum+=1
    return data

def xml_to_df(file,scale=1.5,targets=["1","2"]):
    toDF = {}
    data = parse_xml(file,scale=scale)
    #print(data[targets[1]])
    #print(list(zip(*data[targets[1]])))
    for target in targets:
        if len(data[target]) > 0:
            df = pd.DataFrame(list(zip(*data[target]))).T
            #print(df)
            df.columns = ["X","Y"]
            #print(df)
            toDF[target] = df        
            #df = pd.DataFrame(data[target],columns=["X","Y"])
            #toDF[target] = df
    return toDF

def open_cluster_file(file):
    with open(file) as f:
        X = []
        Y = []
        for line in f:
            boxes = line.split("=")
            for box in boxes:
                x,y,size = box.split("-")
                X.append(x)
                Y.append(y)
        df = pd.DataFrame(list(zip(X,Y)),columns=["BX","BY"])
        df = df.apply(pd.to_numeric)
        #print(df)
        return df
    
def find_unclustered_distance(cells,clusters,target,blength = 150):
    uXs = []
    uYs = []
    uMD = []
    data = clusters
    if target in cells:
        prolif = cells[target][["X","Y"]]
        if len(prolif) > 0:
            for index, row in prolif.iterrows():
                c = row
                indexes = data.loc[((data["BX"]<c["X"])
                                   & ((data["BX"] + blength) >= c["X"])
                                   & (data["BY"]<c["Y"])
                                   & ((data["BY"] + blength) >= c["Y"])
                                   )].index
                if len(indexes) == 0:
                    Xs = data["BX"] - c["X"]
                    Ys = data["BY"] - c["Y"]
                    dist = np.sqrt((Xs**2) + (Ys**2))
                    #c["min_dist"] = np.min(dist)
                    uMD.append(np.min(dist))
                    uXs.append(c["X"])
                    uYs.append(c["Y"])
    
    unclust = pd.DataFrame(list(zip(*[uXs,uYs,uMD])),columns=["X","Y","minDistance"])
    #print(outcl)
    #unclust = pd.concat(outcl)
    return unclust

def unclustered_distance_main(cellpath="G:\\lab_files\\cell_counts\\best\\XMLs\\",clusterpath="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\input_cluster_files\\novel_outlier\\microglia\\combined\\",cell="microglia",target="1",scale=1.5):
    cellfiles = os.listdir(cellpath)
    distances = []
    for file in cellfiles:
        #print(file)
        header,ID,day,cond,sec = file.split("_")
        sec = sec[:-4] #strip .txt
        clfile = ID+"_"+day+"_"+cond+"_"+cell+"_"+sec+".txt"
        clusters = open_cluster_file(clusterpath+clfile)
        #print(clusters.sort_values("BX"))
        cells = xml_to_df(cellpath+file,scale=scale)
        if (not (clusters.empty)) and (target in cells):
            minD = find_unclustered_distance(cells,clusters,target)
            minD["cell"] = [cell]*len(minD)
            minD["ID"] = [ID]*len(minD)
            minD["day"] = [day]*len(minD)
            minD["treatment"] = [cond]*len(minD)
            minD["section"] = [sec]*len(minD)
            distances.append(minD)
    return pd.concat(distances, ignore_index=True)
    
def plot_histogram(Ys,colours=["green"],labels=["default"],line=False):
    
    mu = 0
    sigma = 1
    fig = plt.figure()
    fig2 = plt.figure()
    ax = fig.add_subplot(111)
    ax2 = fig2.add_subplot(111)
    for Y, c, l in zip(Ys,colours,labels):
        n, bins, patches = ax.hist(Y, 20, normed=1, facecolor=c, alpha=0.75, label=l)
        #y = mlab.normpdf(bins, mu, sigma)
        
        ax2.plot(bins,list(n)+[0], c, label = l)
        #print(len(n))
        #print(len(bins))
    #ax2.axvline(x=212,color="black")
    handles, labels = ax.get_legend_handles_labels()
    legend = ax.legend()
    plt.xlabel("nearest cluster")
    plt.ylabel('Probability')
    plt.grid(True)
    plt.show()
#def find_nearest_distance(unclust,clusters):

#min_distances = unclustered_distance_main()

#print(min_distances)

_1D = min_distances.loc[min_distances["day"]=="1D","minDistance"]
_4D = min_distances.loc[min_distances["day"]=="4D","minDistance"]
_7D = min_distances.loc[min_distances["day"]=="7D","minDistance"]
_10D = min_distances.loc[min_distances["day"]=="10D","minDistance"]

plot_histogram([_1D,_4D,_7D,_10D],colours=["green","red","blue","black"],labels=["1D","4D","7D","10D"])



```

```python
import os
import pandas as pd
import numpy as np

def add_cluster_data(threshold_type="novel_outlier",cltype="proximal",d="C:\\Users\\joey\\Documents\\LabWork\\Staining\\immunofluorescense\\Gfap_Iba1_Ki67\\GFAP_Iba1_ki67_batch_output\\output_cluster_files\\",cell="microglia"):
    d_file=d+threshold_type+"\\"+cell+"\\"+cltype+"\\data\\"
    print("adding: %s" %(d_file))
    mice = {}
    files = os.listdir(d_file)
    #print(files)
    for f in files:
        ms, day, cond, cell, sec = f.split("_")
        if sec.endswith(".txt"):
            sec = sec[:-4]
        if cell in mice:
            if day in mice[cell]:
                if ms in mice[cell][day]:
                    if sec in mice[cell][day][ms]:
                        mice[cell][day][ms][sec][cond]=pd.read_csv(d_file+f, delimiter="\t")
                    else:
                        mice[cell][day][ms][sec] = {}
                        mice[cell][day][ms][sec][cond]=pd.read_csv(d_file+f, delimiter="\t")
                else:
                    mice[cell][day][ms] = {}
                    mice[cell][day][ms][sec] = {}
                    mice[cell][day][ms][sec][cond]=pd.read_csv(d_file+f, delimiter="\t")
            else:
                mice[cell][day] = {}
                mice[cell][day][ms] = {}
                mice[cell][day][ms][sec] = {}
                mice[cell][day][ms][sec][cond]=pd.read_csv(d_file+f, delimiter="\t")
        else:
            mice[cell]={}
            mice[cell][day] = {}
            mice[cell][day][ms] = {}
            mice[cell][day][ms][sec] = {}
            mice[cell][day][ms][sec][cond]=pd.read_csv(d_file+f, delimiter="\t")
    return mice

def merge_mouse(ms,threshold=False,cond="FUS",_return=["whole","cluster"]):
    whole = []
    cluster = []
    for sec in ms:
        if threshold:
            whole.append(ms[sec][cond].iloc[1])
            if len(ms[sec][cond]) > 2:
                #print("accessed")
                cluster.append(ms[sec][cond].iloc[3])
        else:
            whole.append(ms[sec][cond].iloc[0])
            if len(ms[sec][cond]) > 2:
                cluster.append(ms[sec][cond].iloc[2])
    #print(cluster)
    whole = pd.DataFrame(whole)
    cluster = pd.DataFrame(cluster)
    if ("whole" in _return) and ("cluster" in _return):
        print("Both")
        return whole, cluster
    elif "whole" in _return:
        return whole
    else:
        return cluster
    
def calculate_measure(mice,operation="Mean",tosubtract=None,_distal=False,cell="microglia",day="1D",cond="FUS",cluster=True,return_ID=False):
    results=[]
    IDs = []
    for ms in mice[cell][day]:
        IDs.append(ms)
        #todo: sum IntDen cluster intDen divide by cluster Area
        if operation == "Mean":
            #calculate distal from whole - combined cluster
            if _distal:
                merged = merge_mouse(mice[cell][day][ms],threshold=True,cond=cond,_return=["whole"])
                tosub = merge_mouse(mice[cell][day][ms],threshold=True,cond=cond,_return=["cluster"])
                IntDen_y = 0
                Area_y = 0
                if len(tosub) > 0:
                    IntDen_y = np.sum(tosub["IntDen"])
                    Area_y = np.sum(tosub["Area"])
                IntDen_x = np.sum(merged["IntDen"])
                Area_x = np.sum(merged["Area"])
                IntDen = IntDen_x - IntDen_y
                Area = Area_x - Area_y
                results.append(IntDen/Area)
            elif tosubtract:
                _return = ["whole"]
                #calculate proximal from combined - focal
                if cluster:
                    _return = ["cluster"]
                merged = merge_mouse(mice[cell][day][ms],threshold=True,cond=cond,_return=_return)
                if len(merged) > 0:
                    #print(merged)
                    IntDen_x = np.sum(merged["IntDen"])
                    Area_x = np.sum(merged["Area"])
                    tosub = merge_mouse(tosubtract[cell][day][ms],threshold=True,cond=cond,_return=_return)
                    #print(tosub)
                    if len(tosub) > 0:
                        #print("length tosub > 0")
                        IntDen_y = np.sum(tosub["IntDen"])
                        Area_y = np.sum(tosub["Area"])
                    else:
                        IntDen_y = 0
                        Area_y = 0
                        
                    IntDen = IntDen_x - IntDen_y
                    Area = Area_x - Area_y
                    
                    #print("IntDenX :%s, IntDenY: %s, AreaX: %s, AreaY: %s, ___ IntDen: %s, Area: %s" %(Area_x, IntDen_x, Area_y, IntDen_y, IntDen, Area))
                    if Area > 0:
                        results.append(IntDen/Area)
                    else:
                        results.append(0)
                else:
                    results.append(0)
            else:
                _return = ["whole"]
                if cluster:
                    _return = ["cluster"]
                merged = merge_mouse(mice[cell][day][ms],threshold=True,cond=cond,_return=_return)
                if len(merged) > 0:
                    IntDen = np.sum(merged["IntDen"])
                    Area = np.sum(merged["Area"])
                    results.append(IntDen/Area)
                else:
                    results.append(0)
        
        elif operation == "%Area":
            if _distal:
                #whole fore - cluster fore / whole back - cluster back
                fore_c = merge_mouse(mice[cell][day][ms],threshold=True,cond=cond,_return=["cluster"])
                back_c = merge_mouse(mice[cell][day][ms],threshold=False,cond=cond,_return=["cluster"])
                fore_w = merge_mouse(mice[cell][day][ms],threshold=True,cond=cond,_return=["whole"])
                back_w = merge_mouse(mice[cell][day][ms],threshold=False,cond=cond,_return=["whole"])
                if len(fore_c) > 0:
                    fore = np.sum(fore_w["Area"]) - np.sum(fore_c["Area"])
                    back = np.sum(back_w["Area"]) - np.sum(back_c["Area"])
                else:
                    fore - np.sum(fore_w["Area"])
                    back = np.sum(back_w["Area"])
                percArea = fore/back
                results.append(percArea)
            elif tosubtract:
                _return = ["whole"]
                if cluster:
                    _return = ["cluster"]
                fore_x = merge_mouse(mice[cell][day][ms],threshold=True,cond=cond,_return=_return)
                back_x = merge_mouse(mice[cell][day][ms],threshold=False,cond=cond,_return=_return)
                fore_y = merge_mouse(tosubtract[cell][day][ms],threshold=True,cond=cond,_return=_return)
                back_y = merge_mouse(tosubtract[cell][day][ms],threshold=False,cond=cond,_return=_return)
                if len(fore_x) > 0:
                    if len(fore_y) > 0:
                        fore = np.sum(fore_x["Area"]) - np.sum(fore_y["Area"])
                        back = np.sum(back_x["Area"]) - np.sum(back_y["Area"])
                    else:
                        fore = np.sum(fore_x["Area"])
                        back = np.sum(back_x["Area"])
                    percArea = fore/back
                else:
                    percArea = 0
                results.append(percArea)
            else:
                _return = ["whole"]
                if cluster:
                    _return = ["cluster"]
                fore = merge_mouse(mice[cell][day][ms],threshold=True,cond=cond,_return=_return)
                if len(fore) > 0:
                    back = merge_mouse(mice[cell][day][ms],threshold=False,cond=cond,_return=_return)
                    percArea = np.sum(fore["Area"])/np.sum(back["Area"])
                else:
                    percArea = 0
                results.append(percArea)
                    
        elif operation == "cluster_area":
            if _distal:
                merged = merge_mouse(mice[cell][day][ms],threshold=False,cond=cond,_return=["whole"])
                tosub = merge_mouse(mice[cell][day][ms],threshold=False,cond=cond,_return=["cluster"])
                Area_x = np.sum(merged["Area"])
                if len(tosub) > 0:
                    Area_y = np.sum(tosub["Area"])
                else:
                    Area_y = 0
                Area = Area_x - Area_y
                results.append(Area)
                
            elif tosubtract:
                _return = ["whole"]
                #calculate proximal from combined - focal
                if cluster:
                    _return = ["cluster"]
                merged = merge_mouse(mice[cell][day][ms],threshold=False,cond=cond,_return=_return)
                tosub = merge_mouse(tosubtract[cell][day][ms],threshold=False,cond=cond,_return=_return)
                if len(merged) > 0:
                    Area_x = np.sum(merged["Area"])
                    if len(tosub) > 0:
                        Area_y = np.sum(tosub["Area"])
                    else:
                        Area_y = 0
                    Area = Area_x - Area_y
                else:
                    Area = 0
                if ms == "JS10":
                    print(merged)
                    print(tosub)
                    print(Area_y)
                    print(Area_x)
                    print(Area)
                results.append(Area)
            else:
                _return = ["whole"]
                if cluster:
                    _return = ["cluster"]
                merged = merge_mouse(mice[cell][day][ms],threshold=False,cond=cond,_return=_return)
                if len(merged) > 0:
                    Area = np.sum(merged["Area"])
                else:
                    Area = 0
                results.append(Area)
    
    if return_ID:
        return results, IDs
    else:
        return results
    
def write_cluster_file(file,data,path="C:\\Users\\joey\\Documents\\LabWork\\Staining\\immunofluorescense\\Gfap_Iba1_Ki67\\GFAP_Iba1_ki67_batch_output\\Analysis\\",Transpose=None,columns=["noFUS","FUS","Focal","Proximal","Distal","IDs"],header=None,method="novel_outlier",cell="microglia"):
    full_path = path+method+"\\"+cell+"\\"+file
    with open(full_path,'w') as w:
        if (header == None) == False:
            for v in header:
                w.write(v)
                w.write("\t")
            w.write("\n")
        if Transpose == None:
            tablelen = len(data[0])
            for col in columns:
                w.write(col)
                w.write("\t")
            w.write("\n")
            for row in range(0,tablelen,1):
                #print("row number %s" %(row))
                for col in data:
                    #print("column of data: %s" %(col))
                    w.write(str(col[row]))
                    w.write("\t")
                w.write("\n")
        else:
            coLen = Transpose[1] - Transpose[0]
            tablelen = len(data[0])
            #iter through table columnns
            for ID in range(0,tablelen,1):#note each iter of data represents a new mouse
                #print("ID is: %s" %(ID))
                for treat in range(0,coLen,1):#each iter of collen is a new treatment
                    w.write(str(data[treat][ID]))
                    #print("i is %s" %(i))
                    w.write("\t")
                    w.write(str(treat))
                    w.write("\t")
                    w.write(columns[treat])
                    for c in data[Transpose[1]:]:
                        w.write("\t")
                        w.write(str(c[ID]))
                    w.write("\n")
                        #print(c)
                        #print("mouse is: %s" %(c[ID]))

def analyze_all(_dir="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\output_cluster_files\\",
                days=["1D","4D","7D","10D","30D"],method=["novel_outlier"],cells=["microglia"],cond="FUS",base=True,
                output_dir="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\Gfap_Iba1_analysis\\",
                measures=["Mean","cluster_area","%Area"]):
    for me in method:
        for cell in cells:
            if cell == "microglia":
                focal =add_cluster_data(cltype="focal",threshold_type=me,d=_dir,cell=cell)
                combined =add_cluster_data(cltype="combined",threshold_type=me,d=_dir,cell=cell)
                
                for day in days:
                    towrite = []
                    cols = []
                    if base:
                        #base analysis
                        #towrite += [nofus_x,fus_x,nofus_cA,fus_cA, nofus_pA, fus_pA]
                        #cols += ["whole_noFUS_mean","whole_FUS_mean","whole_noFUS_Area","whole_FUS_Area","whole_noFUS_%Area","whole_FUS_%Area"]
                        
                        if "Mean" in measures:
                            nofus_x, IDs = calculate_measure(combined,operation="Mean",day=day,cell=cell,cluster=False,cond="noFUS",return_ID=True)
                            fus_x = calculate_measure(combined,operation="Mean",day=day,cell=cell,cluster=False,cond="FUS")
                            towrite += [nofus_x,fus_x]
                            cols += ["whole_noFUS_mean","whole_FUS_mean"]
                        
                        if "cluster_area" in measures:
                            nofus_cA, IDs = prox_cA = calculate_measure(combined,operation="cluster_area",day=day,cell=cell,cluster=False,cond="noFUS",return_ID=True)
                            fus_cA = calculate_measure(combined,operation="cluster_area",day=day,cell=cell,cluster=False,cond="FUS")
                            towrite += [nofus_cA,fus_cA]
                            cols+= ["whole_noFUS_Area","whole_FUS_Area"]
                        
                        if "%Area" in measures:
                            nofus_pA, IDs = calculate_measure(combined,operation="%Area",day=day,cell=cell,cluster=False, cond="noFUS", return_ID=True)
                            fus_pA = calculate_measure(combined,operation="%Area",day=day,cell=cell,cluster=False)
                            towrite += [nofus_pA, fus_pA]
                            cols += ["whole_noFUS_%Area","whole_FUS_%Area"]
                    
                    #condition specific analysis
                    #Mean
                    #towrite += [focal_x,prox_x,dist_x,foc_cA,prox_cA,dist_cA,foc_pA,prox_pA,dist_pA,IDs]
                    #cols += ["focal_mean","proximal_mean","distal_mean","focal_cluster_area","proximal_cluster_area","distal_cluster_area","focal_%Area","proximal_%Area","distal_%Area","Mouse_ID"]
                    
                    if "Mean" in measures:
                        focal_x = calculate_measure(focal,operation="Mean",day=day,cell=cell,cond=cond)
                        prox_x = calculate_measure(combined,tosubtract=focal,operation="Mean",day=day,cell=cell,cond=cond)
                        dist_x = calculate_measure(combined,operation="Mean",_distal=True,day=day,cell=cell,cond=cond)
                        towrite += [focal_x,prox_x,dist_x]
                        cols += ["focal_mean","proximal_mean","distal_mean"]
                    
                    #cluster_Area
                    if "cluster_area" in measures:
                        foc_cA = calculate_measure(focal,operation="cluster_area",day=day,cell=cell,cond=cond)
                        prox_cA = calculate_measure(combined,tosubtract=focal,operation="cluster_area",day=day,cell=cell,cond=cond)
                        dist_cA = calculate_measure(combined,operation="cluster_area",_distal=True,day=day,cell=cell,cond=cond)
                        towrite += [foc_cA,prox_cA,dist_cA]
                        cols += ["focal_cluster_area","proximal_cluster_area","distal_cluster_area"]
                        
                    #percent_Cell_Area
                    if "%Area" in measures:
                        foc_pA = calculate_measure(focal,operation="%Area",day=day,cell=cell,cond=cond)
                        prox_pA = calculate_measure(combined,operation="%Area",tosubtract=focal,day=day,cell=cell,cond=cond)
                        dist_pA = calculate_measure(combined,operation="%Area",_distal=True,day=day,cell=cell,cond=cond)
                        towrite += [foc_pA,prox_pA,dist_pA,IDs]
                        cols += ["focal_%Area","proximal_%Area","distal_%Area"]
                    
                    towrite+= [IDs]
                    cols += ["mouse_IDs"]
                    
                    write_cluster_file(day+"_"+cond+"_"+str(measures)+"_microglia_data_table.txt",towrite,columns=cols,Transpose=None,cell=cell,method=me,path=output_dir)
                    write_cluster_file(day+"_"+cond+"_"+str(measures)+"_microglia_data_transposed_table.txt",towrite,columns=cols,Transpose=(0,len(cols)-1),header=["Mean","Treatment","Str_Treatments","ID"],cell=cell,method=me,path=output_dir)
            
            if cell == "astrocyte":
                #focal =add_cluster_data(cltype="focal",threshold_type=thresh_type,d=_dir,cell=cell)
                proximal =add_cluster_data(cltype="proximal",threshold_type=me,d=_dir,cell=cell)
                for day in days:
                    towrite = []
                    cols = []
                    if base:
                        #base analysis
                        if "Mean" in measures:
                            nofus_x, IDs = calculate_measure(proximal,operation="Mean",day=day,cell=cell,cluster=False,cond="noFUS",return_ID=True)
                            fus_x = calculate_measure(proximal,operation="Mean",day=day,cell=cell,cluster=False,cond="FUS")
                            towrite += [nofus_x,fus_x]
                            cols += ["whole_noFUS_mean","whole_FUS_mean"]
                            
                        if "cluster_area" in measures:
                            nofus_cA, IDs = prox_cA = calculate_measure(proximal,operation="cluster_area",day=day,cell=cell,cluster=False,cond="noFUS",return_ID=True)
                            fus_cA = prox_cA = calculate_measure(proximal,operation="cluster_area",day=day,cell=cell,cluster=False,cond="FUS")
                            towrite += [nofus_cA,fus_cA]
                            cols += ["whole_noFUS_Area","whole_FUS_Area"]
                        
                        if "%Area" in measures:
                            nofus_pA, IDs = calculate_measure(proximal,operation="%Area",day=day,cell=cell,cluster=False, cond="noFUS", return_ID=True)
                            fus_pA = calculate_measure(proximal,operation="%Area",day=day,cell=cell,cluster=False)
                            towrite += [nofus_pA, fus_pA]
                            cols += ["whole_noFUS_%Area","whole_FUS_%Area"]
                    #condition specific analysis
                    #Mean
                    if "Mean" in measures:
                        prox_x = calculate_measure(proximal,operation="Mean",day=day,cell=cell,cond=cond)
                        dist_x = calculate_measure(proximal,operation="Mean",_distal=True,day=day,cell=cell,cond=cond)
                        #print(prox_x)
                        towrite += [prox_x,dist_x]
                        cols += ["proximal_mean","distal_mean"]
                    #cluster_Area
                    if "cluster_area" in measures:
                        prox_cA = calculate_measure(proximal,operation="cluster_area",day=day,cell=cell,cond=cond)
                        dist_cA = calculate_measure(proximal,operation="cluster_area",_distal=True,day=day,cell=cell,cond=cond)
                        towrite += [prox_cA,dist_cA]
                        cols += ["proximal_cluster_area","distal_cluster_area"]
                    #percent_Cell_Area
                    if "%Area" in measures:
                        prox_pA = calculate_measure(proximal,operation="%Area",day=day,cell=cell,cond=cond)
                        dist_pA = calculate_measure(proximal,operation="%Area",_distal=True,day=day,cell=cell,cond=cond)
                        towrite += [prox_pA,dist_pA,IDs]
                        cols += ["proximal_%Area","distal_%Area"]
                    towrite+= [IDs]
                    cols += ["mouse_IDs"]
                    #if base:
                    #    towrite += [nofus_x,fus_x,nofus_cA,fus_cA, nofus_pA, fus_pA]
                    #    cols += ["whole_noFUS_mean","whole_FUS_mean","whole_noFUS_Area","whole_FUS_Area","whole_noFUS_%Area","whole_FUS_%Area"]
                    #towrite += [prox_x,dist_x,prox_cA,dist_cA,prox_pA,dist_pA,IDs]
                    #cols += ["proximal_mean","distal_mean","proximal_cluster_area","distal_cluster_area","proximal_%Area","distal_%Area","ID"]
                    
                    write_cluster_file(day+"_"+cond+"_"+str(measures)+"_astrocyte_data_table.txt",towrite,columns=cols,Transpose=None,cell=cell,method=me,path=output_dir)
                    write_cluster_file(day+"_"+cond+"_"+str(measures)+"_astrocyte_data_transposed_table.txt",towrite,columns=cols,Transpose=(0,len(cols)-1),header=["Mean","Treatment","Str_Treatments","ID"],cell=cell,method=me,path=output_dir)

    
#ms = focal["microglia"]["7D"]["JS16"]
#print(ms)
#wh,cl = merge_mouse(ms,threshold=False)
#print(len(wh))
#print(wh)
#print(len(cl))
#print(cl)

#day="10D"
#_day_=day+"_"

_dir = "F:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\output_cluster_files\\"
out_dir = "F:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\Gfap_Iba1_analysis\\"
#thresh_type = "novel_outlier"
#cell = "microglia"
#day = "10D"


analyze_all(cells=["astrocyte"],measures=["Mean"],_dir=_dir,output_dir=out_dir)
analyze_all(cells=["astrocyte"],measures=["cluster_area"],_dir=_dir,output_dir=out_dir)
analyze_all(cells=["astrocyte"],measures=["%Area"],_dir=_dir,output_dir=out_dir)


analyze_all(cells=["microglia"],measures=["Mean"],_dir=_dir,output_dir=out_dir)
analyze_all(cells=["microglia"],measures=["cluster_area"],_dir=_dir,output_dir=out_dir)
analyze_all(cells=["microglia"],measures=["%Area"],_dir=_dir,output_dir=out_dir)

print("done")

#write_cluster_file("test.txt",[noFUS,fus,IDs],columns=["noFUS","FUS","ID"],Transpose=None)
#write_cluster_file("test_T.txt",[noFUS,fus,IDs],columns=["noFUS","FUS","ID"],Transpose=(0,2),header=["Mean","Treatment","Str_Treatments","ID"])
        
```

```python
import os
import pandas as pd
import numpy as np

def read_skeletons(path="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\output_cluster_files\\novel_outlier\\microglia\\", types=["focal","proximal","distal"]):
    results = []
    for t in types:
        p = path+t+"\\skeleton\\"
        files = os.listdir(p)
        raw_data = []
        branch_data = []
        raw_cols = ["ID","Day","Treatment","Cell","Section","#Branches","#Junctions","#EndPoints"]
        branch_cols = ["ID","Day","Treatment","Cell","Section","BranchLength"]
        for f in files:
            #print(f)
            ID, day, treat, cell, sec, dtype = f.split("_")
            dtype = dtype[:-4] #remove .txt
            data = pd.read_csv(p+"\\"+f,delimiter="\t")
            
            if dtype == "rawInfo":
                numBranches, numJunctions, numEnds = sum_raw_data(data)
                raw_data.append([ID,day,treat,cell,sec,numBranches, numJunctions, numEnds])
            elif dtype == "branchInfo":
                branchLen = sum_branch_data(data)
                branch_data.append([ID,day,treat,cell,sec,branchLen])
        raw = pd.DataFrame(raw_data,columns=raw_cols)
        branch = pd.DataFrame(branch_data,columns=branch_cols)
        result = pd.merge(raw, branch, how='outer', on=["ID","Day","Treatment","Cell","Section"])
        #print(result)
        results.append(result)
    return results
            
            
def sum_branch_data(data):
    return np.sum(data["Branch length"])
    
def sum_raw_data(data):
    nBranch = np.sum(data["# Branches"])
    nJunc = np.sum(data["# Junctions"])
    nEnds = np.sum(data["# End-point voxels"])
    return nBranch, nJunc, nEnds

def read_soma_file(path="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\output_cluster_files\\novel_outlier\\microglia\\",types=["focal","proximal","distal"],extrapath=""):
    results = []
    cols = ["ID","Day","Treatment","Cell","Section","SomaNum","SomaSize"]
    for t in types:
        p = path+t+extrapath+"\\counts\\"
        files = os.listdir(p)
        counts = []
        for f in files:
            print(f)
            ID, day, treat, cell, sec = f.split("_")
            sec = sec[:-4]
            data = pd.read_csv(p+"\\"+f,delimiter="\t")
            count, soma_size = sum_counts(data)
            counts.append([ID,day,treat,cell,sec,count, soma_size])
        c = pd.DataFrame(counts,columns=cols)
        results.append(c)
    return results

def sum_counts(data):
    count = len(data)
    sum_size = np.sum(data["Area"])
    return count, sum_size
    
            
def merge_soma_skeleton(soma,skeleton):
    return pd.merge(soma, skeleton, how='outer', on=["ID","Day","Treatment","Cell","Section"])

def analyze_complexity(data, toSubtract = [], add=False):
    cols = ["ID","Day","Treatment","Cell","SomaNum","SomaSize","BranchLength","#Branches","#Junctions","#EndPoints"]
    row = []
    for ID in np.unique(data["ID"]):
        d = data.loc[data["ID"]==ID]
        total_bl = np.sum(d["BranchLength"])
        total_nB = np.sum(d["#Branches"])
        total_nJ = np.sum(d["#Junctions"])
        total_nEP = np.sum(d["#EndPoints"])
        total_sArea = np.sum(d["SomaSize"])
        total_sNum = np.sum(d["SomaNum"])
        if len(toSubtract)>0:
            sbtr = toSubtract.loc[toSubtract["ID"]==ID]
            sbtr_bl = np.sum(sbtr["BranchLength"])
            sbtr_nB = np.sum(sbtr["#Branches"])
            sbtr_nJ = np.sum(sbtr["#Junctions"])
            sbtr_nEP = np.sum(sbtr["#EndPoints"])
            sbtr_sArea = np.sum(sbtr["SomaSize"])
            sbtr_sNum = np.sum(sbtr["SomaNum"])
            if add:
                total_bl = total_bl + sbtr_bl
                total_nB = total_nB + sbtr_nB
                total_nJ = total_nJ + sbtr_nJ
                total_nEP = total_nEP + sbtr_nEP
                total_sArea = total_sArea + sbtr_sArea
                total_sNum = total_sNum + sbtr_sNum
            else:
                print("total SomaNum: %s, focal SomaNum : %s" %(total_sNum,sbtr_sNum))
                total_bl = total_bl - sbtr_bl
                total_nB = total_nB - sbtr_nB
                total_nJ = total_nJ - sbtr_nJ
                total_nEP = total_nEP - sbtr_nEP
                total_sArea = total_sArea - sbtr_sArea
                total_sNum = total_sNum - sbtr_sNum
                
        row.append([d["ID"].iloc[0],d["Day"].iloc[0],d["Treatment"].iloc[0],d["Cell"].iloc[0],total_sNum,total_sArea,total_bl,total_nB, total_nJ, total_nEP])
    summary = pd.DataFrame(row, columns=cols)
    summary["meanBranchLength"] = summary["BranchLength"]/summary["SomaNum"]
    summary["meanNumBranches"] = summary["#Branches"]/summary["SomaNum"]
    summary["meanNumJunctions"] = summary["#Junctions"]/summary["SomaNum"]
    summary["meanEndPoints"] = summary["#EndPoints"]/summary["SomaNum"]
    summary["meanSomaSize"] = summary["SomaSize"]/summary["SomaNum"]
    return summary
        

    
def summarize_skeletons(skelPath="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\output_cluster_files\\novel_outlier\\microglia\\",
                        countPath="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\output_cluster_files\\novel_outlier\\microglia\\",
                        outPath="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\skeleton_analysis_out\\",
                        types=[("focal","FUS"),("distal","FUS"),("unclustered","noFUS")] ,
                        days=["1D","4D","7D","10D"],
                        sepOut=False,
                        cellCountExtraPath="",
                        subtract_pair=[[("combined","FUS"),("focal","FUS"),"proximal",False],[("distal","FUS"),("unclustered","FUS"),"distal-unclustered",True]]
                       ):
    outTypes=types
    result = []
    for t_ in types:
        print(t_)
        t = t_[0]
        cond = t_[1]
        s = read_skeletons(path=skelPath,types=[t])
        print("skeletons read")
        c = read_soma_file(path=countPath, types=[t],extrapath=cellCountExtraPath)
        print("somas read")
        r = merge_soma_skeleton(s[0],c[0])
        #print(r)
        if sepOut:
            result = []
        for d in days:
            r_ = r.loc[(r["Day"]==d) & (r["Treatment"]==cond)]
            r_ = analyze_complexity(r_)
            r_["Type"] = [t]*len(r_)
            result.append(r_)
        if ( sepOut ):
            _r_ = pd.concat(result, ignore_index=True)
            outFN = t + "_" + cond + ".txt"
            _r_.to_csv(outPath+outFN,sep="\t")
    print("done type")
    if subtract_pair:
        for pair in subtract_pair:
            print("===========")
            print(pair[0])
            print("===========")
            Ttype = pair[0][0]
            Tcond = pair[0][1]
            Ts = read_skeletons(path=skelPath,types=[Ttype])
            Tc = read_soma_file(path=countPath, types=[Ttype])
            Tr = merge_soma_skeleton(Ts[0],Tc[0])
            
            Stype = pair[1][0]
            Scond = pair[1][1]
            Ss = read_skeletons(path=skelPath,types=[Stype])
            Sc = read_soma_file(path=countPath, types=[Stype])
            Sr = merge_soma_skeleton(Ss[0],Sc[0])
            if sepOut:
                result = []
            outTypes.append(pair[2])
            for d in days:
                Tr_ = Tr.loc[(Tr["Day"]==d) & (Tr["Treatment"]==Tcond)]
                Sr_ = Sr.loc[(Sr["Day"]==d) & (Sr["Treatment"]==Scond)]
                r_ = analyze_complexity(Tr_, toSubtract=Sr_, add=pair[3])
                print("--------------------")
                print(pair[2])
                print("--------------------")
                r_["Type"] = [pair[2]]*len(r_)
                result.append(r_)
            if ( sepOut ):
                _r_ = pd.concat(result, ignore_index=True)
                outFN = t + "_" + cond + ".txt"
                _r_.to_csv(outPath+outFN,sep="\t")
    if(sepOut == False):
        _r_ = pd.concat(result, ignore_index=True)
        outFN = str(outTypes) + "_" + cond + ".txt"
        _r_.to_csv(outPath+outFN,sep="\t")
        
#summarize_skeletons()

summarize_skeletons(skelPath="G:\\lab_files\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\",
                    countPath="G:\\lab_files\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\",
                    outPath="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\skeleton_analysis_out\\",
                    types=[("output_whole","noFUS"),("output_whole","FUS")],
                    days = ["1D","4D","7D","10D"],
                    cellCountExtraPath="\\cell_counts",
                    subtract_pair=[],
                    sepOut=False
                   )
        
'''
focal_skeletons, distal_skeletons = read_skeletons(types=["focal","distal"])
#print(result)
#print(sum(result.loc[result["Treatment"] == "FUS",["#EndPoints"]].values))
#print(sum(result.loc[result["Treatment"] == "noFUS",["#EndPoints"]].values))

focal_counts, distal_counts = read_soma_file(types=["focal","distal"])

#print(skeletons)
#print(counts)
focal_data = merge_soma_skeleton(focal_counts,focal_skeletons)
distal_data = merge_soma_skeleton(distal_counts,distal_skeletons)


_1D_f = focal_data.loc[(focal_data["Day"]=="1D") & (focal_data["Treatment"]=="FUS")]
_1D_d = distal_data.loc[(distal_data["Day"]=="1D") & (distal_data["Treatment"]=="FUS")]

print(analyze_complexity(_1D_f))
print("######")
print(analyze_complexity(_1D_d))
'''
'''
_4D = data.loc[(data["Day"]=="4D") & (data["Treatment"]=="FUS")]
_7D = data.loc[(data["Day"]=="7D") & (data["Treatment"]=="FUS")]
_10D = data.loc[(data["Day"]=="10D") & (data["Treatment"]=="FUS")]

print(analyze_complexity(_1D))
print("######")
print(analyze_complexity(_4D))
print("######")
print(analyze_complexity(_7D))
print("######")
print(analyze_complexity(_10D))
'''

        
```

```python
import os
import pandas as pd
import numpy as np

def read_nnd_files(path="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\output_cluster_files\\novel_outlier\\microglia\\",cltype="combined"):
    p = path + cltype + "\\NND"
    NNDs = []
    for f in os.listdir(p):
        ID, day, treat, cell, sec, dtype = f.split("_")
        dtype = dtype[:-4] #remove .txt
        data = pd.read_csv(p+"\\"+f,delimiter="\t")
        data = data.rename(columns={'---':'NND'})
        data = data.rename(columns={' ':'posID'})
        data["posID"] = data["posID"].astype(int)
        data["ID"] = [ID]*len(data)
        data["day"] = [day]*len(data)
        data["treatment"] = [treat]*len(data)
        data["cell"] = [cell]*len(data)
        data["section"] = [sec]*len(data)
        data["cluster"] = [cltype]*len(data)
        NNDs.append(data)
    return pd.concat(NNDs,ignore_index=True)

def read_cell_roi_files(path="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\output_cluster_files\\novel_outlier\\microglia\\",cltype="combined",extraPath=""):
    p = path + cltype + extraPath + "\\counts"
    cellPos = []
    for f in os.listdir(p):
        ID, day, treat, cell, sec = f.split("_")
        sec = sec[:-4]
        data = pd.read_csv(p+"\\"+f,delimiter="\t")
        data = data.rename(columns={' ':'posID'})
        data["posID"] = data["posID"].astype(int)
        data["ID"] = [ID]*len(data)
        data["day"] = [day]*len(data)
        data["treatment"] = [treat]*len(data)
        data["cell"] = [cell]*len(data)
        data["section"] = [sec]*len(data)
        data["cluster"] = [cltype]*len(data)
        cellPos.append(data)
    return pd.concat(cellPos, ignore_index=True)

def read_cluster_files(path="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\output_cluster_files\\novel_outlier\\microglia\\",
                       cltypes=["combined","focal","proximal","distal","unclustered"],read_nnd_file=True,read_roi_file=True,cellCountExtraPath=""):
    DFs = {}
    if read_nnd_file:
        nndDFs = []
        for cl in cltypes:
            print(cl)
            nndDFs.append(read_nnd_files(path=path, cltype=cl))
        DFs["NND"] = pd.concat(nndDFs, ignore_index=True)
    if read_roi_file:
        roiDFs = []
        for cl in cltypes:
            roiDFs.append(read_cell_roi_files(path=path, cltype=cl, extraPath=cellCountExtraPath))
        DFs["ROI"] = pd.concat(roiDFs, ignore_index=True)
        
    return DFs

def subtract_data(refdata, subtractdata, refType="combined", subtractType="focal",mergeOn=["XM","YM","X","Y"],identifiers=["cell","ID","day","section","treatment"],dropSubtract=True,resultType=None):
    ref = refdata.loc[refdata["cluster"]==refType]
    sub = subtractdata.loc[subtractdata["cluster"]==subtractType]
    sub["mergeOn"] = [True]*len(sub)
    
    toDrop = pd.merge(ref,sub[mergeOn+identifiers+["mergeOn"]],how="outer",on=mergeOn+identifiers)
    if resultType:
        toDrop["cluster"] = [resultType]*len(toDrop)
    if dropSubtract:
        result = toDrop.loc[~(toDrop["mergeOn"]==True)]
    else:
        result = toDrop.loc[toDrop["mergeOn"]==True]
    return result

def add_data(df1,df2,df1clust="distal",df2clust="unclustered",resultType="distal-unclustered"):
    DFs=[]
    #print(df1["cluster"])
    DFs.append(df1.loc[df1["cluster"] == df1clust ])
    DFs.append(df2.loc[df2["cluster"] == df2clust ])
    result = pd.concat(DFs, ignore_index=True)
    if resultType:
        result["cluster"] = [resultType] * len(result)
    return result
    #print()
      
def summarize_days(data,treat,cltype,days=["1D","4D","7D","10D"],combine=None):
    summary=[]
    sumcols = ["day","ID","treatment","cluster","mean_NND"]
    for day in days:
        d = data.loc[data["day"]==day]
        for ID in np.unique(d["ID"]):
            nnds = d.loc[(d["ID"]==ID) & (d["treatment"]==treat) & (d["cluster"]==cltype)]
            if combine:
                cb = d.loc[(d["ID"]==ID) & (d["treatment"]==combine[0]) & (d["cluster"]==combine[1])]
                nnds = pd.concat([nnds,cb])
            #nnds = nnds.dropna()
            sumrow = [day,ID,treat,cltype,np.mean(nnds["NND"])]
            summary.append(sumrow)
    return pd.DataFrame(summary,columns=sumcols)
    

def analyze_nnd(path="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\output_cluster_files\\novel_outlier\\microglia\\",
              outpath="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\nnd_analysis_out\\",
              cltypes=["combined","focal","distal","unclustered"],
                final=[["focal","FUS"],["unclustered","noFUS"],["distal","FUS"],["unclustered","FUS"]],
              subtractPairs=[["combined","focal","proximal","FUS"]],
              addPairs = [["distal","unclustered","FUS","distal-unclustered"]],
                cellCountExtraPath = ""
                
             ):
    print()
    DFs = read_cluster_files(path=path, cltypes=cltypes,cellCountExtraPath=cellCountExtraPath)
    roiData = DFs["ROI"]
    nndData = DFs["NND"]
        
    for cl in final:
        #print( DFs["NND"].loc[DFs["NND"]["treatment"]=="noFUS"] )
        sum_nnd = summarize_days(DFs["NND"],cl[1],cl[0])
        outFN = str(cl)+"_NND_summary.txt"
        sum_nnd.to_csv(outpath+outFN,sep="\t")
        print(sum_nnd)
        #sum_nnd.write_csv()
    
    
    for pair in subtractPairs:
        refN = pair[0]
        subN = pair[1]
        resultN = pair[2]
        cond = pair[3]
        
        #eliminate the subtractType cells from the reference datatype
        sub_rois = subtract_data(roiData,roiData,refType=refN,subtractType=subN,mergeOn=["XM","YM","X","Y"],identifiers=["cell","ID","day","section","treatment"],dropSubtract=True,resultType=None)
        
        #use subtracted reference ROIs to retrieve reference cell positions.
        sub_nnd = subtract_data(nndData,sub_rois,refType=refN,subtractType=refN, mergeOn=["posID"],identifiers=["cell","ID","day","section","treatment"], dropSubtract=False,resultType=resultN)
        
        sum_sub_nnd = summarize_days(sub_nnd,cond,resultN)
        outFN = resultN+"_"+cond+"_NND_summary.txt"
        sum_sub_nnd.to_csv(outpath+outFN,sep="\t")
        print(sum_sub_nnd)
        
    for pair in addPairs:
        refN = pair[0]
        subN = pair[1]
        resultN = pair[2]
        cond = pair[3]
        
        add_nnd = add_data(nndData,nndData,df1clust=pair[0],df2clust=pair[1],resultType=pair[3])
        sum_add_nnd = summarize_days(add_nnd,pair[2],cltype=pair[3])
        outFN = resultN+"_"+cond+"_NND_summary.txt"
        sum_add_nnd.to_csv(outpath+outFN,sep="\t")
        
        #print(sum_add_nnd)
        
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                
'''   
    
#comb = read_cell_roi_files()
#focal = read_cell_roi_files(cltype="focal")

#proximal_pos = subtract_data(comb,focal)

#print(proximal_pos)
#print(_)
#nnd_files = read_nnd_files()

#print(proximal_pos["posID"].values)
#proximal_pos["posID"] = proximal_pos["posID"].astype(int)
#print(proximal_pos["posID"])

#print(_["posID"])

proximal_pos["posID"] = proximal_pos["posID"].astype(int)
nnd_files["posID"] = nnd_files["posID"].astype(int)

nnd_prox = subtract_data(nnd_files,proximal_pos,refType="combined",subtractType="combined",mergeOn=["posID"],dropSubtract=False,resultType="proximal")

nnd_prox_summarized = summarize_days(nnd_prox,"FUS","proximal")
nnd_combined_summarized = summarize_days(nnd_files,"FUS","combined")

print(nnd_prox)
print(nnd_prox_summarized)
print(nnd_combined_summarized)
#print(nnd_prox)



#print(nnd_files)

#print(np.round(nnd_prox["posID"].values))

    
#_ = read_nnd_files()
#__ = summarize_days(_,"FUS","combined")

#print(__)
'''

#analyze_nnd()

analyze_nnd(path="G:\\lab_files\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\",
           outpath="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\nnd_analysis_out\\",
           cltypes=["output_whole"],
           final=[["output_whole","noFUS"],["output_whole","FUS"]],
           subtractPairs=[],
           addPairs=[],
           cellCountExtraPath="\\cell_counts"
           )

print("done")
    
```

```python
import os
import pandas as pd
import numpy as np


#overlap combined file:
# microglia cluster, astrocyte cluster, overlap cluster -- microgliaOnly = microglia - overlap, astrocyteOnly = astrocyte - overlap
#
def read_overlaps(path="G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\GFAP_Iba1_ki67_output\\output_cluster_files\\novel_outlier\\cluster_overlap\\", types=["combined","microglia","astrocyte"]):
    cols = ["ID","Day","Treatment","Section","Type","Overlapping","MicrogliaOnly","AstrocyteOnly"]
    re = []
    for t in types:
        p = path+t
        files = os.listdir(p)
        for f in files:
            #print(f)
            ID, day, treat, sec = f.split("_")
            sec = sec[:-4] #remove .txt
            data = pd.read_csv(p+"\\"+f,delimiter="\t")
            if len(data)>1:
                micro = data.loc[0,"Area"]
                astro = data.loc[1,"Area"]
                merged = data.loc[2,"Area"]
                
                micro = micro - merged
                astro = astro - merged
                               
            else:
                diff = data.loc[0,"Area"]
                if t == "microglia":
                    micro = diff
                    astro = 0
                elif t == "astrocyte":
                    astro = diff
                    micro = 0
                merged = 0
            _new = pd.DataFrame([[ID,day,treat,sec,t,merged,micro,astro]],columns=cols)
            re.append(_new)
    return pd.concat(re,ignore_index=True)

def iter_samples(data,key="ID",treatments=["FUS","noFUS"]):
    samples = np.unique(data[key])
    for s in samples:
        for t in treatments:
            yield data.loc[(data[key]==s)&(data["Treatment"]==t)]
        
        
def summarize_mice(overlaps):
    summary = []
    
    IDcols = np.setdiff1d(overlaps.columns.values,np.array(["Overlapping","MicrogliaOnly","AstrocyteOnly"]))
    for ms in iter_samples(overlaps):
        #print(ms)
        d = list(ms[["Overlapping","MicrogliaOnly","AstrocyteOnly"]].sum(axis=0).values)
        #print(list(d.values))
        IDvals = ms[IDcols]
        if len(IDvals) > 0:
            IDvals = list(IDvals.iloc[0].values)
            summary.append(IDvals + d)
    summarized_overlaps = pd.DataFrame(summary,columns=list(IDcols)+["Overlapping","MicrogliaOnly","AstrocyteOnly"])
    summarized_overlaps["perc_Overlap"] = (summarized_overlaps["Overlapping"] / (summarized_overlaps["Overlapping"] + summarized_overlaps["MicrogliaOnly"] + summarized_overlaps["AstrocyteOnly"]) )
    summarized_overlaps["perc_AstrocyteOnly"] = (summarized_overlaps["AstrocyteOnly"] / (summarized_overlaps["Overlapping"] + summarized_overlaps["MicrogliaOnly"] + summarized_overlaps["AstrocyteOnly"]) )
    summarized_overlaps["perc_MicrogliaOnly"] = (summarized_overlaps["MicrogliaOnly"] / (summarized_overlaps["Overlapping"] + summarized_overlaps["MicrogliaOnly"] + summarized_overlaps["AstrocyteOnly"]) )
    return summarized_overlaps

overlap = read_overlaps()
#print(overlap)

summarized_overlaps = summarize_mice(overlap)

summarized_overlaps.to_csv("G:\\lab_files\\imageJ_macro_working_directory\\Gfap_Iba1_ki67\\Gfap_Iba1_analysis\\novel_outlier\\cluster_overlap\\cluster_overlap.txt","\t")

'''
overlap = overlap.loc[(overlap["Type"]=="combined") & (overlap["Treatment"]=="FUS")]
astrocyte = overlap.loc[(overlap["Type"]=="astrocyte") & (overlap["Treatment"]=="FUS")]
microglia = overlap.loc[(overlap["Type"]=="microglia") & (overlap["Treatment"]=="FUS")]

for d in ["1D","4D","7D","10D"]:
    ov = overlap.loc[overlap["Day"]==d]
    m = microglia.loc[microglia["Day"]==d]
    a = astrocyte.loc[astrocyte["Day"]==d]
    
    #print(ov["Overlapping"])
    print(float(np.sum(ov["Overlapping"])))
    
    print(np.sum(m["Difference"]))
    print(np.sum(a["Difference"]))
    
    print(float((np.sum(ov["Overlapping"]) + np.sum(m["Difference"]) + np.sum(a["Difference"]) )))
    percOV = float(np.sum(ov["Overlapping"])) / float( ( (np.sum(ov["Overlapping"]) + np.sum(ov["Difference"]) + np.sum(m["Difference"]) + np.sum(a["Difference"]) )) )
    print("day: %s, overlap: %0.3f" %(d,percOV))
'''
```

```python
import pandas as pd
from sklearn import svm
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import math
import matplotlib.pyplot as plt
import copy

'''
This function takes a dataset, and partitions it by k conjoined dataspaces
'''
def k_partition(X,k=4,_returnjoined=False,spatialcoordinates=["BX","BY"],test_index=0):
    partitions = []
    splitlen = int(len(X)/k)
    X = X.sort_values(by=spatialcoordinates, ascending=True)
    for i in range(0,k):
        start = splitlen*i
        stop = start + splitlen
        partitions.append(X.iloc[start:stop])
    if _returnjoined:
        if k > 2:
            test = partitions.pop(test_index)
            #print("pop! length of partitions = %s" %(len(partitions)))
            train = pd.concat(partitions,ignore_index=False)
            partitions = [train,test]
        elif k == 2:
            test = partitions.pop(test_index)
            train = partitions[0]
            partitions = [train,test]
        elif k == 1:
            partitions = [X,X]
        #print(len(partitions))
    return partitions

def remove_background(X,dropna=True):
    data = X
    data = data.sort_values(by="Area",ascending=False)
    #print("length of data: %s" %len(data))
    bg_row = data.iloc[0]
    bg = bg_row["Mean"]
    rm = data.iloc[2:]
    rmdata = rm.loc[rm["Mean"]>bg,rm.columns.values]
    #print("length of rmdata: %s" %len(rmdata))
    if dropna:
        rmdata = rmdata.dropna()
    return rmdata

def score_outlier_detection_w_cross_validation(sec,_nu=0.15,_gamma='auto',_kernel='rbf',features=["Mean","IntDen","D"],spatialfeatures=["BX","BY"],verbose=False,cross_validation=4,_min=[1,21,3], test_fus=False):
    if verbose:
        print("parameters -- nu: %s, gamma: %s, kernel: %s, features: %s, -- cross validation: %s" %(_nu, _gamma, _kernel, features, cross_validation))
    #remove_background
    nf = sec.data["noFUS"]
    #print(nf)
    
    nf = remove_background(nf)
    
    #scale   
    scaler = StandardScaler()
    scaler.fit(nf[features])
    nf[features] = scaler.transform(nf[features])
    #nf = pd.DataFrame(_nf,columns=_nf.columns.values)
    all_features = features+spatialfeatures
    if test_fus:
        f = sec.data["FUS"]
        f = remove_background(f)
        f[features] = scaler.transform(f[features])
        
    #novel outlier
    estimator = svm.OneClassSVM(nu=_nu,gamma=_gamma,kernel=_kernel)
    
    #dbscan iterators
    _eps = math.sqrt(2.0*sec.boxsize**2)/1.5
    scores = {}
    if cross_validation == 0:
        cross_validation_iters = 1
    else:
        cross_validation_iters = cross_validation
    for i in range(0,cross_validation_iters):
        if verbose:
            print("cross validation %s" %(i))
        if test_fus:
            train = nf
            test = f
        else:
            train, test = k_partition(nf[all_features],k=cross_validation,_returnjoined=True,test_index=i)
            if verbose:
                print("train data: %s, test data: %s" %(len(train), len(test)))
        #print(train)
        estimator.fit(train[features])
        test["outlier"] = estimator.predict(test[features])
        if verbose:
            print("target: %s, percent outliers: %s" %(_nu,len(test.loc[test["outlier"]==-1])/(len(test.loc[test["outlier"]==-1])+len(test.loc[test["outlier"]==1]))))
        tocluster = test.loc[test["outlier"]==-1]
        for j in range(_min[0],_min[1],_min[2]):
            score=[]
            clusterer=DBSCAN(min_samples=j,eps=_eps)
            clusters = clusterer.fit_predict(tocluster[spatialfeatures])
            #print(clusters)
            incluster = (clusters > -1).sum()
            #score.append(incluster)
            scores[j] = incluster
    return scores

    
#internal method
def add_score(scores,nu,gamma,m,ID,score):
    if nu in scores:
        if gamma in scores[nu]:
            if m in scores[nu][gamma]:
                if ID in scores[nu][gamma][m]:
                    scores[nu][gamma][m][ID].append(score)
                else:
                    scores[nu][gamma][m][ID] = [score]
            else:
                scores[nu][gamma][m] = {}
                scores[nu][gamma][m][ID] = [score]
        else:
            scores[nu][gamma]={}
            scores[nu][gamma][m] = {}
            scores[nu][gamma][m][ID] = [score]
    else:
        scores[nu] = {}
        scores[nu][gamma]={}
        scores[nu][gamma][m] = {}
        scores[nu][gamma][m][ID] = [score]
    return scores

#internal method
def sum_scores(scores):
    xyzv = [[],[],[],[]]
    for nu in scores:
        for gamma in scores[nu]:
            for m in scores[nu][gamma]:
                tosum=[]
                for ID in scores[nu][gamma][m]:
                    tosum.append(np.sum(scores[nu][gamma][m][ID])) #sum score for a mouse
                scores[nu][gamma][m]=np.sum(tosum)
                xyzv[0].append(nu)
                xyzv[1].append(gamma)
                xyzv[2].append(m)
                xyzv[3].append(np.sum(tosum))
    return xyzv,scores

#a messs!!!!
def summarize_scores(scores):
    ID_scores={}
    for ID in scores:
        for s in scores[ID]:
            for nu in scores[ID][s]:
                for gamma in scores[ID][s][nu]:
                    for m in scores[ID][s][nu][gamma]:
                        score = scores[ID][s][nu][gamma][m]
                        ID_scores = add_score(ID_scores,nu,gamma,m,ID,score)
    #print(ID_scores)
    summary = sum_scores(ID_scores)
    return summary

def plot_scores(summary):
    fig = plt.figure()
    ax = fig.add_subplot(111,projection='3d')
    
    
def grid_search(mice,cell="microglia",day="1D",_nu=[0.05,0.30,0.05],_gamma=[0.1,0.6,0.1],_min=[10,26,5],cross_validation=2,verbose=False, test_fus=False, features=["Mean","IntDen","D"]):
    scores={}
    for ID in mice[cell][day]:
        secs = get_mouse_sections(mice,cell,day,ID)
        scores[ID] = {}
        for s in secs:
            print("============ %s %s ===========" %(ID,s))
            scores[ID][s]={}
            sec = copy.deepcopy(secs[s])
            nu_iters = int( (_nu[1] - _nu[0]) / _nu[2] ) + 1
            gamma_iters = int( (_gamma[1] - _gamma[0]) / _gamma[2] ) + 1
            for n in range(0,nu_iters):
                nu = _nu[0]+n*_nu[2]
                scores[ID][s][nu] = {}
                for g in range(0,gamma_iters):
                    gamma = _gamma[0]+g*_gamma[2]
                    print("iteration: nu: %s, gamma: %s" %(nu,gamma))
                    scores[ID][s][nu][gamma] = score_outlier_detection_w_cross_validation(sec,verbose=verbose,_nu=nu,_min=_min,_gamma=gamma,cross_validation=cross_validation,test_fus=test_fus, features=features)
    return scores
    
    #print(scores)
    #scores = summarize_scores(scores)
    #print(scores)

s = "sec2"
cell = "microglia"
day="1D"
ID="JS1"

#secs = get_mouse_sections(mice,cell,day,ID)
#sec = copy.deepcopy(secs[s])
#scores = score_outlier_detection_w_cross_validation(sec,verbose=True,_nu=0.1,_min=[10,26,2],_gamma='auto',cross_validation=2)
#print(scores)
#print(summarize_scores(scores))



#scores = grid_search(mice, test_fus=False, cross_validation=1, cell="astrocyte", verbose=True, day="4D")
#xyzv_4D, summary_4D = summarize_scores(scores)
def grid_search_days(days=["1D","4D","7D","10D"],features=["Mean","Area","IntDen","D"], cell="astrocyte", 
                     _nu=[0.05,0.30,0.05],_gamma=[0.1,0.6,0.1],_min=[10,26,5]):
    grids = []
    cols = ["nu","gamma","min","score"]
    t_cols = ["nu","gamma","min","treated_score"]
    for d in days:
        scores = grid_search(mice, test_fus=False, cross_validation=1, cell=cell, verbose=False, day=d, features=features, _nu=_nu, _gamma=_gamma, _min=_min)
        xyzv, summary_un = summarize_scores(scores)
        _un = pd.DataFrame(list(zip(*xyzv)),columns = cols)
        scores = grid_search(mice, test_fus=True, cross_validation=1, cell=cell, verbose=False, day=d, features=features, _nu=_nu, _gamma=_gamma, _min=_min)
        xyzv, summary_un = summarize_scores(scores)
        _t = pd.DataFrame(list(zip(*xyzv)),columns = t_cols)
        _x = pd.merge(_un, _t, how='left', on=['nu', 'gamma', 'min'])
        _x["day"] = [d]*len(_x)
        #zeroes = _x.loc[_x["score"] == 0]
        grids.append(_x)
    return pd.concat(grids,ignore_index=True)

def combine_days(data):
    _nu = []
    _gamma = []
    _min = []
    score = []
    for n in np.unique(data["nu"]):
        for g in np.unique(data["gamma"]):
            for m in np.unique(data["min"]):
                d = float(np.sum(data.loc[(data["nu"] == n) & (data["gamma"] == g) & (data["min"] == m),["treated_score"]]))
                #print(d)
                _nu.append(n)
                _gamma.append(g)
                _min.append(m)
                score.append(d)
    return pd.DataFrame(list(zip(*[_nu,_gamma,_min,score])),columns=["nu","gamma","min","treated_score"])

def get_zeroes(_x):
    zeroes = _x.loc[_x["score"] == 0]
    return zeroes

#"Mean","IntDen","D","microglia_soma_roundness","microglia_soma_mean_area"
features = ["Mean","IntDen","D","microglia_soma_roundness","microglia_soma_mean_area"]
#features = ["Mean","IntDen","D","Area"]

#_nu=[0.1,0.26,0.03],_gamma=[0.1,0.6,0.1],_min=[15,27,2],

#_all = grid_search_days(cell="microglia",features=features,
#                       _nu=[0.05,0.21,0.02],_gamma=[0.05,0.31,0.05],_min=[10,26,2], days=["1D","4D","7D","10D"])
print(_all.sort_values(by="treated_score", ascending=False))
zeroes = get_zeroes(copy.copy(_all))
zeroes = zeroes.loc[zeroes["gamma"] < 0.2]
#print(zeroes.sort_values(by="treated_score",ascending=False))

for d in np.unique(zeroes["day"]):
    _ = zeroes.loc[zeroes["day"]==d]
    print(_.sort_values(by="treated_score",ascending=False).iloc[0:10])
    
print(combine_days(zeroes).sort_values(by="treated_score",ascending=False))


ast_features=["Mean","IntDen","D","Area"]
#_all_ast = grid_search_days(cell="astrocyte",features=ast_features,
#                       _nu=[0.05,0.21,0.02],_gamma=[0.05,0.31,0.05],_min=[10,26,2], days=["1D","4D","7D","10D"])
print(_all.sort_values(by="treated_score", ascending=False))
zeroes_ast = get_zeroes(copy.copy(_all_ast))
zeroes_ast = zeroes_ast.loc[zeroes_ast["gamma"] < 0.2]
#print(zeroes.sort_values(by="treated_score",ascending=False))

for d in np.unique(zeroes["day"]):
    _ast = zeroes.loc[zeroes["day"]==d]
    print(_ast.sort_values(by="treated_score",ascending=False).iloc[0:10])
    
print(combine_days(zeroes_ast).sort_values(by="treated_score",ascending=False))
        #print(zeroes.sort_values(by="treated_score", ascending=False))
#xyvz

#print(xyzv)
```
